```python
#a) Implement a FIFO data structure in python.
class FIFO(object):
    def __init__(self):
        self.items = []

    def is_empty(self):
        return self.items == []

    def insert(self, item):
        self.items.insert(0,item)

    def size(self):
        return len(self.items)

    def remove(self):
        return self.items.pop()


#b) Use the implemented FIFO data structure to implement breadth first search (BFS) algorithm for Tree 1 in python. The starting nodes for Tree 1 is ‘F’.
def bfs(tree, start):
    visited = []
    queue = FIFO()

    queue.insert(start)
    visited.append(start)

    while queue.size() > 0:
        vertex = queue.remove()
        for i in tree[vertex]:
            if i not in visited:
                queue.insert(i)
                visited.append(i)

    return visited


Tree1 = {
    'F': ['L', 'W', 'J'],
    'L': ['K'],
    'W': ['N', 'E'],
    'J': ['M'],
    'K': ['A'],
    'N': ['S'],
    'A': [],
    'S': [],
    'E': [],
    'M': []
}

bfs(Tree1, 'F')

#c)Print out Visited nodes, Queue stack, and pint out results of BFS.
print('Visited: ', bfs(Tree1, 'F'))
#define queue items as list
queue_items=['F', 'L', 'W', 'J', 'K', 'N', 'E', 'M', 'A', 'S']
print('Queue: ', queue_items)

#d) Draw and visualize BFS algorithm applied on Tree1 using networkx library showing Visited, Queue and print as outputs.
import networkx as nx

G = nx.Graph()

for i in Tree1:
    G.add_node(i)
    for j in Tree1[i]:
        G.add_edge(i, j)

nx.draw(G)
```
